## **Linear data vs Non Linear data :**




```
# Linear data --> predictor and response variable has follows a "straight line pattern" relationship

# Non Linear data --> predictor and response variable has not follows a "straight line pattern" relationship (Needs some curve or bend or complex shapes)

```



## **Why Polynomial Regression ?**







```
# Non linear data has relationship but it cant be  represented by the linear model Effectively.

# So here we are using  "Polynomial Regression" to add polynomial terms(higher powers) of the predictor variable to create "polynomial features"

# By Using the Polynomial features , model becomes more flexible , can capture curved or complex relationships present in the data

# Polynomial regression can capture model curved relationships effectively.

Note :

      # Adding more features more flexible will increase in regression line
      
          (Higher degree of polynomials may leads to overfitting so be aware of degrees as well as the smaller value of degree  tries to underfit so find optimum value )

      # Here we select best fit curve that minimises the residual b.w predictor and response data points
```

## **Polynomial Degree:**






```
# The Polynomial Degree is use to adding the power to predictor feature to create polynomial features (x,x^2,x^3 ....X^n)
```
**Degrees**


```
# Degree 1 (linear regression)     : y = b0 + b1*x

# Degree 2 (quadratic polynomial)  : y = b0 + b1*x + b2*x^2

# Degree 3 (Cubic polynomial)      : y = b0 + b1*x + b2*x^2 + b3 * x^3

# Degree 4 (quartic polynomial )   : y = b0 + b1*x + b2*x^2 + b3*x^3 + b4*x^4

# Degree 5 (quintic polynomial)    : y = b0 + b1*x + b2*x^2 + b3*x^3 + b4*x^4 + b5*x^5

# Degree n (general polynomial)    : Y = b0 + b1*x + b2*x^2 + ... + bn*x^n
```



## **Example**

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({'sqrt':['100','200','300','400','500','600'],'Price':[25000,100,1040,40000,200,3700]})

df
````

````python
x = df[['sqrt']]

y = df['Price']
````

````python
model = LinearRegression().fit(x,y)
````

````python
plt.scatter(x['sqrt'],y,color='green')

y_pred = model.predict(x)

plt.plot(x['sqrt'],y_pred,color='r')

plt.show()
````

### **Handling Non Linear Data With Polynomial Regression**

````python
from sklearn.preprocessing  import PolynomialFeatures
````

````python
# Adding Degress

for i in [1,2,3,4,5]:

  poly = PolynomialFeatures(degree=i)      # Degree

  x_poly = poly.fit_transform(x)           # X data tranformed into polynomial 'x' feature

  model = LinearRegression().fit(x_poly,y)

  plt.scatter(x['sqrt'],y,color='green')  # Normal data

  y_pred = model.predict(x_poly)

  print()
  print('SSE :',r2_score(y,y_pred))
  print()

  plt.plot(x['sqrt'],y_pred,color='r')    # Predicted data

  plt.title(f'Polynomial regression Graph {i}')

  plt.xlabel('Square feet')

  plt.ylabel('Price')

  plt.show()
````

### **Conclusion**

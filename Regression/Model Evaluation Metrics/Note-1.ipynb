# **Regression Evaluation Metrics :**

## **Mean Absolute Error**


























**Goal Of Metric :**

```
# Assess the average magnitude of errors in the model's predictions compared to the actual values.
                          (Measuring the average absolute prediction error)

```

**Why is it used ?**


```
# It provides a measure of the typical error in the model's predictions, allowing us to understand how much the model deviates from the actual data points on average.

```

**Formula :**



```
  MAE = Σ|actual - predicted|/n

  # Results in a value representing the average magnitude of errors, with the same unit as the data.
```

**Example :**



```
Example: Mean Absolute Error (Housing Price Prediction)

Actual Prices: [10, 20, 30]
Predicted Prices: [12, 18, 32]

# Calculate MAE for this example:

MAE = (|10-12| + |20-18| + |30-32|) / 3    =>  (2 + 2 + 2) / 3    => 2

MAE = 2

# Interpretation:

- The MAE value  2 indicates that  on average, the model's predictions have an absolute error of approximately 2 units when compared to the actual housing prices.


```
**Note**


```
# Lower MAE Indicates that the model predictions are closer to actual values
  --> so error or residuals less.


# Higher MAE Indicates that the model predictions are not closer  to actual values
  --> so error or residuals Higher.
```



**Conclusion :**


```
#  MAE quantifies the average magnitude of prediction errors, making it a useful metric for assessing the model's overall accuracy in predicting continuous numerical values.

```
**Disadvantage :**



```
# Disadvantage: MAE treats all errors equally, which might not be ideal if large errors should be penalized more. Eg : outlier
```

**When to use MAE Metric ?**

```
Use MAE when you want a metric that's easy to interpret and all errors should be treated equally.
```


## **Mean Squared Error:**



**Goal Of Metric**


```
# Assess the average magnitude of "squared errors" in the model's predictions compared to the actual values.
                          (Measuring the average squared prediction error)
```
**Why is it used ?**

```
# It provides a measure of the "typical squared error" in the model's predictions, allowing us to understand how much the model deviates from the actual data points on average while giving more weight to larger errors.
```
**Formula :**



```
# MSE = Σ(actual - predicted)^2/n

# Results in a value representing the average magnitude of squared errors, with the squared units of the data.
```

**Example :**



```
Example: Mean Squared Error (Housing Price Prediction)

Actual Prices: [10, 20, 30]
Predicted Prices: [12, 18, 32]

# Calculate MSE for this example:

MSE = ((10-12)^2 + (20-18)^2 + (30-32)^2) / 3
MSE = (4 + 4 + 4) / 3
MSE = 4

# Interpretation:

- The MSE value of 4 indicates that, on average, the model's predictions have a squared error of approximately 4 units when compared to the actual housing prices.
```
**Note :**


```
# Lower MSE indicates that the model predictions are closer to actual values, with smaller squared errors.
  --> so errors or residuals are less spread out.

# Higher MSE indicates that the model predictions have larger squared errors and are not as close to actual values, with more spread-out errors.
  --> so errors or residuals are more spread out.
```

**Disadvantages**


```
# By squaring the errors , the scaling of the original errors will change. This error will be treated by rmse by taking root it will bring backs to the original error scale.
```

**When to use MSE Metric ?**

```
Use MSE when you want a metric that penalizes larger errors more than MAE..
```




## **Root Mean Squared Error :**






**Goal Of Metric :**


```
# Assess the average magnitude of "squared errors" in the model's predictions compared to the actual values, similar to MSE, but with the added step of taking the square root to bring the units back to the original scale.
```
**Why is it used ?**

```
# RMSE provides a measure of the "typical squared error" in the model's predictions, just like MSE, but it's particularly useful when you want to report errors in the "original units" of the data rather than squared units.
```
**Formula :**



```
#  RMSE = √(Σ(actual - predicted)^2/n)

  # Results in a value representing the average magnitude of squared errors, with the units of the data with root.

```
**Example :**



```
Example: Root Mean Squared Error (Housing Price Prediction)

Actual Prices: [10, 20, 30]
Predicted Prices: [12, 18, 32]

# Calculate RMSE for this example:

RMSE = √(((10-12)^2 + (20-18)^2 + (30-32)^2) / 3)
RMSE = √((4 + 4 + 4) / 3)
RMSE = √(4)
RMSE = 2

# Conclusion:

- The RMSE value of 2 indicates that, on average, the model's predictions have a squared error of approximately 4 units when compared to the actual housing prices. However, because RMSE takes the square root, it brings the units back to the original scale, so the typical error is approximately 2 units in the original data units
```

**Note :**



```
# Lower RMSE indicates that the model predictions are closer to actual values, with smaller squared errors in the original data units.
  --> so errors or residuals are less spread out in the original units.

# Higher RMSE indicates that the model predictions have larger squared errors in the original data units and are not as close to actual values.
  --> so errors or residuals are more spread out in the original units.

```

**When to use RMSE metric ?**


```
# Use RMSE when you want the errors in the same units as the target variable and still penalize larger errors.
```
